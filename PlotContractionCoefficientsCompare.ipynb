```python
import glob, os
import numpy as np
import matplotlib.pyplot as plt
os.chdir('Results')
```

```python
parts=2
i=0
coefficients=[0.2,0.4,0.6,0.1,0.1,1]
plt.style.use('ggplot')
    
for file in glob.glob("CIFAR10*.npy"):
    print(file)
    TestLoss=np.load(file)
    [M,N]=TestLoss.shape
    TestLossEpoches=TestLoss[:,N-1]
    if i<parts:
        plt.plot(TestLossEpoches, lw=1.5)
    else:
        plt.plot(TestLossEpoches,'--', lw=1.5)
    i+=1   

WithoutContraction=np.loadtxt('CIFAR10128TestConvergenceChanges.csv')
plt.plot(WithoutContraction[0:9], lw=1.5)
plt.xlabel('Epoches')
plt.ylabel('Loss')

plt.legend(tuple(coefficients))
plt.savefig('ContractionCoefficientsCompare_CIFAR10.png',dpi=600)
```
