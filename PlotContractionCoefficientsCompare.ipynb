```python
import glob, os
import numpy as np
import matplotlib.pyplot as plt
os.chdir('Results')
```

```python
parts=2
i=0
coefficients=[601,600.5,1001,1000.5,1401,1405]
plt.style.use('ggplot')

flielist=['CIFAR10[60, 1]TestConvergenceChanges.npy', 'CIFAR10[60, 0.5]TestConvergenceChanges.npy', 
          'CIFAR10[100, 1]TestConvergenceChanges.npy','CIFAR10[100, 0.5]TestConvergenceChanges.npy',
         'CIFAR10[140, 1]TestConvergenceChanges.npy','CIFAR10[140, 0.5]TestConvergenceChanges.npy']
    
for i in range(6):
    TestLoss=np.load('BatchSizeComp/'+flielist[i])
    TestLossEpoches=np.mean(TestLoss,1)
    if i<parts:
        plt.plot(TestLossEpoches, lw=1.5)
    else:
        plt.plot(TestLossEpoches,'--', lw=1.5)
    i+=1   


plt.xlabel('Epoches')
plt.ylabel('Loss')

plt.legend(tuple(coefficients))
plt.savefig('ContractionCoefficientsCompare_CIFAR10.png',dpi=600)
```
